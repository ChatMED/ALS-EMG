```python
import pandas as pd
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,roc_auc_score
import json
from random import sample
import os
import mne
import pandas as pd
from tqdm import tqdm
```

```python
from sklearn.model_selection import train_test_split
from itertools import combinations
import math
import numpy as np
```

```python
pd.options.mode.chained_assignment = None
```

```python
results=pd.DataFrame(columns=["Dataset","Accuracy","Precision","Recall","F1-Score","ROC-AUC"])
path="../LLM/eeg_datasets/control_vs_als_dataset"
for ds in os.listdir(path):
    train_df=pd.read_csv(os.path.join(path,ds,"train_final.csv"))
    test_df=pd.read_csv(os.path.join(path,ds,"test_final.csv"))

    x_train,y_train=train_df.drop(columns=["PatientId","Class"]),train_df["Class"]
    x_test,y_test=test_df.drop(columns=["PatientId","Class"]),test_df["Class"]

    model=XGBClassifier()
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)

    results.loc[len(results)]=[ds,accuracy_score(y_test,y_pred),precision_score(y_test,y_pred),recall_score(y_test,y_pred),f1_score(y_test,y_pred),roc_auc_score(y_test,y_pred)]

results
```

```python
accuracy=results["Accuracy"].mean()
precision=results["Precision"].mean()
recall=results["Recall"].mean()
f1score=results["F1-Score"].mean()
rocauc=results["ROC-AUC"].mean()

print(accuracy,precision,recall,f1score,rocauc)
```
